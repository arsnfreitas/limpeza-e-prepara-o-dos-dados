### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae

```python
sinasc_raw[sinasc_raw['DTNASC'] == '2019-04-19']['SEXO'].value_counts()
```

```python
sinasc_raw[sinasc_raw['DTNASC'] == '2019-04-19'].groupby(['ESCMAE', 'SEXO'])['PESO'].mean().unstack()
```

### 3. Qual o municipio onde nasceram menos bebês em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw['munResNome'].value_counts(ascending = True).head()
```

```python
menos_bebes = sinasc_raw[sinasc_raw['munResNome'] == 'Castanheiras']
menos_bebes.groupby('munResNome').agg({'IDADEMAE': ['mean','max', 'min'], 
                                                                      'IDADEPAI': ['mean','max', 'min']})
```
